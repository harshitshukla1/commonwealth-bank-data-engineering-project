```python
import pandas as pd
```

```python
df = pd.read_excel('supermarket_transactions.xlsx')
```

```python
df.head()
```

```python
df.columns
```

```python
#Across locations, how many apples were purchased in cash?
apple_cash = df[
    (df['product_name'].str.lower() == 'apple') &
    (df['payment_method'].str.lower() == 'cash')
]
```

```python
apple_quantity = apple_cash['quantity'].sum()

print(apple_quantity)
```

```python
apple_total_cash = apple_cash['total_amount'].sum()

print(apple_total_cash)
```

```python
#How much total cash was spent on these apples?
apple_total_cash = apple_cash['total_amount'].sum()

print(apple_total_cash)
```

```python
#Across all payment methods, how much money was spent at the Bakershire store location by non-member customers?
```

```python
bakershire_nonmember = df[
    (df['store'] == 'Bakershire') &
    (df['customer_type'] == 'non-member')
]
```

```python
bakershire_total = bakershire_nonmember['total_amount'].sum()

print(bakershire_total)
```

```python
print("Final Answers")
print("----------------------")
print("1. Apples purchased in cash:", apple_quantity)
print("2. Total cash spent on apples: $", apple_total_cash)
print("3. Bakershire non-member spending: $", bakershire_total)
```

```python
summary = pd.DataFrame({
    'Question': [
        'Apples purchased in cash',
        'Total cash spent on apples',
        'Bakershire non-member spending'
    ],
    'Answer': [
        apple_quantity,
        apple_total_cash,
        bakershire_total
    ]
})

summary.to_excel('final_answers.xlsx', index=False)
```
