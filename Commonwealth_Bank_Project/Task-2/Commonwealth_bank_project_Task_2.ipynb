```python
import pandas as pd
```

```python
df = pd.read_excel('mobile_customers.xlsx')
```

```python
df.columns
```

```python
df = df.drop(columns=['Unnamed: 0'])
```

```python
df = df.drop(columns=[
    'name',
    'address',
    'birthdate',
    'credit_card_number',
    'credit_card_security_code',
    'credit_card_expire'
])
```

```python
df['customer_id'] = (
    'CUST_' +
    df['customer_id'].astype(str).str[-4:]
)  #masked customer id column
```

```python
df['username'] = (
    df['username'].astype(str).str[:2] +
    '****'
) #masked username
```

```python
df['email'] = df['email'].apply(
    lambda x: x[:2] + '****@****'
) #masked email
```

```python
bins = [0, 18, 30, 45, 60, 100]

labels = [
    '0-18',
    '19-30',
    '31-45',
    '46-60',
    '60+'
]

df['age_group'] = pd.cut(
    df['age'],
    bins=bins,
    labels=labels
)

df = df.drop(columns=['age'])
```

```python
salary_bins = [0, 30000, 60000, 100000, 200000]

salary_labels = [
    'Low',
    'Medium',
    'High',
    'Very High'
]

df['salary_bracket'] = pd.cut(
    df['salary'],
    bins=salary_bins,
    labels=salary_labels
)

df = df.drop(columns=['salary']) #salary bracket
```

```python
# FINAL ANONYMISED DATASET OUTPUT

print("\nFINAL ANONYMISED DATASET")
print(df.head())

print("\nFINAL COLUMNS")
print(df.columns)
```

```python
df.to_csv(
    'anonymised_mobile_customers.csv',
    index=False
)
```
